```python
house_price_dataset = sklearn.datasets.load_boston() #() means we take 1 insatnce of the data
house_price_dataset
```

```python
house_price_dataset.target
```

```python
#loading the dataset to a pandas dataframe and inputing the target column as well

house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
house_price_dataframe['price'] = pd.Series(house_price_dataset.target)
house_price_dataframe
```

```python
df = house_price_dataframe
```

```python
df.head()
```

```python
#number of rows and columns in the df
df.shape
```

```python
#check for missing values
df.isnull().sum()
```

```python
#stats of data
df.describe()
```

```python
import seaborn as sns
```

```python
#correlation of the data
correlation = df.corr()

#plot the correlation via heatmap
plt.figure(figsize=(8,8))
sns.heatmap(
    correlation, #data
    cmap='Blues', #
    annot=True, #feature names
    annot_kws={'size':8}, #annotation size
    cbar=True, #colorbar be available
    square=True, #value to be in a square
    fmt='.1f' # float values ie 1dp
)
```

```python
#splitting data and target
X = df.drop('price', axis=1)
y = df['price']
```

```python
print(X)
print(y)
```

```python
#splitting data into training data and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2) #splitting data to reproduce code
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#model training
#XGboost is ensemble ie more than 1 DT model.

#loading
model = XGBRegressor()

#fitting data
model.fit(X_train, y_train) #data and label
```

```python
#prediction
#get accuracy
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
#evaluation using r2, mae
score1 = metrics.r2_score(y_train, training_data_prediction) # variance btwn true and predicted
score2 = metrics.mean_absolute_error(y_train, training_data_prediction) # find difference then square

print('r2 error', score1) # be close to 1
print('MAE', score2) # be close to 0
```

```python
#plot error values
plt.scatter(y_train, training_data_prediction)
plt.xlabel('Actual prices')
plt.ylabel('Predicted prices')
plt.title('Actaul price VS Predicted price')
plt.show()
```

```python
#accuracy for test data
test_data_prediction = model.predict(X_test)
```

```python
#evaluate testing data
score1 = metrics.r2_score(y_test, test_data_prediction) # variance btwn true and predicted
score2 = metrics.mean_absolute_error(y_test, test_data_prediction) # find difference then square

print('r2 error', score1) # be close to 1
print('MAE', score2) # be close to 0
```

```python
#plotting the error values
plt.scatter(y_test, test_data_prediction)
plt.xlabel('Actual prices')
plt.ylabel('Predicted prices')
plt.title('Actaul price VS Predicted price')
plt.show()
```
